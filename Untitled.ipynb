### Dataset Return(python import)  
- **`dataset`**
	- **`data`**: Zawiera macierze zestawu danych jako ramki danych **pandas**
		- `ids`: Ramka danych kolumn ID
		- `features`: Ramka danych kolumn cech

		- `original`: Ramka danych zawierająca wszystkie ID, cechy i wartości docelowe
		- `headers`: Lista wszystkich nazw/nagłówków zmiennych
	- **`metadata`**: Zawiera informacje o metadanych zestawu danych
		- Szczegóły znajdują się w sekcji Metadata poniżej
	- **`variables`**: Zawiera szczegóły zmiennych przedstawione w formie tabelarycznej/ramki danych
		- `name`: Nazwa zmiennej
		- `role`: Czy zmienna jest ID, cechą, czy wartością docelową
		- `type`: Typ danych, np. kategorialne, całkowite, ciągłe
		- `demographic`: Wskazuje, czy zmienna reprezentuje dane demograficzne
		- `description`: Krótki opis zmiennej
		- `units`: Jednostki zmiennej dla danych niekategorialnych
		- `missing_values`: Czy w kolumnie zmiennej występują brakujące wartości

## Metadata 
- uci_id: Unikalny identyfikator zestawu danych dla repozytorium UCI
- name: Nazwa
- abstract: Krótki opis zestawu danych
- area: Obszar tematyczny, np. nauki przyrodnicze, biznes
- task: Powiązane zadania uczenia maszynowego, np. klasyfikacja, regresja
- characteristics: Typy zestawów danych, np. wielowymiarowe, sekwencyjne
- num_instances: Liczba wierszy lub próbek
- num_features: Liczba kolumn cech
- feature_types: Typy danych cech
- target_col: Nazwa kolumny(kolumn) docelowych
- index_col: Nazwa kolumny(kolumn) indeksu
- has_missing_values: Czy zestaw danych zawiera brakujące wartości

- year_of_dataset_creation: Rok utworzenia zestawu danych
- dataset_doi: DOI zarejestrowane dla zestawu danych, które linkuje do strony zestawu danych w repozytorium UCI
- creators: Lista nazw twórców zestawu danych
- intro_paper: Informacje o opublikowanym artykule wprowadzającym do zestawu danych
- repository_url: Link do strony internetowej zestawu danych w repozytorium UCI
- data_url: Link do pliku surowych danych
- additional_info: Opisowy tekst swobodny o zestawie danych
- summary: Ogólne podsumowanie
- purpose: W jakim celu stworzono ten zestaw danych?
- funding: Kto sfinansował stworzenie tego zestawu danych?
- instances_represent: Co reprezentują instancje w tym zestawie danych?
- recommended_data_splits: Czy są zalecane podziały danych?
- sensitive_data: Czy zestaw danych zawiera dane, które mogą być w jakikolwiek sposób uważane za wrażliwe?
- preprocessing_description: Czy wykonano jakieś wstępne przetwarzanie danych?
- variable_info: Dodatkowy tekst swobodny z opisem zmiennych
- citation: Prośby o cytowanie/podziękowania
- external_url: URL do zewnętrznej strony zestawu danych. To pole będzie istnieć tylko dla połączonych zestawów danych, tj. nie hostowanych przez UCI.

# 3. Przygotowanie danych

## a) Czyszczenie danych

Najbardziej brakujące kolumny to:

```python
# Sprawdźmy liczbę brakujących wartości w każdej kolumnie
missing_values = data.isnull().sum()

# Procent brakujących danych dla orientacji
missing_percent = (missing_values / len(data)) * 100

# Połączmy wyniki w jedną tabelę
missing_summary = pd.DataFrame({
    "Brakujące wartości": missing_values,
    "Procent braków (%)": missing_percent.round(2)
}).sort_values(by="Brakujące wartości", ascending=False)

missing_summary.head(10)
```

### Niektóre kolumny mają bardzo duży odsetek braków (>30%), więc:
- Kolumny z ponad 30% braków (rbc, rc) — możemy je usunąć lub uzupełnić tylko jeśli są krytyczne.
- Kolumny 10–30% — sensowne do imputacji (np. średnia, mediana, tryb).

### Zacznijmy od:

1. Usunięcia id.

2. Uzupełnienia danych:

   - liczbowych medianą,
   - kategorycznych trybem.
     
3. Potem przejdziemy do wyboru cech i przekształceń. 

```python
# Usuń kolumnę 'id', jeśli istnieje
df_clean = data.drop(columns=["id"], errors="ignore")

# Wybierz kolumny liczbowe i kategoryczne
numeric_cols = df_clean.select_dtypes(include=["float64", "int64"]).columns
categorical_cols = df_clean.select_dtypes(include="object").columns

# Imputacja medianą dla danych liczbowych
num_imputer = SimpleImputer(strategy="median")
df_clean[numeric_cols] = num_imputer.fit_transform(df_clean[numeric_cols])

# Imputacja trybem dla danych kategorycznych
cat_imputer = SimpleImputer(strategy="most_frequent")
df_clean[categorical_cols] = cat_imputer.fit_transform(df_clean[categorical_cols])

# Sprawdź, czy pozostały jakieś braki
print("Brakujące wartości po czyszczeniu:", df_clean.isnull().sum().sum())
```
